```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
df = pd.read_csv(r"C:\Users\aishw\OneDrive\Documents\Python Proj\Amazon-sales-Analysis\data\Amazon Sale Report.csv")

# Show first 5 rows
df.head()
df.info()
df.describe()
```

```python
# ðŸ”¹ 1. Drop useless columns (ignore if already missing)
df = df.drop(columns=["index", "New", "PendingS"], errors="ignore")

# ðŸ”¹ 2. Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)

# ðŸ”¹ 3. Convert postal code to string safely
if 'ship-postal-code' in df.columns:
    df['ship-postal-code'] = df['ship-postal-code'].astype(str)
    df['ship-postal-code'] = df['ship-postal-code'].replace("nan", pd.NA)

# ðŸ”¹ 4. Handle missing values
if 'currency' in df.columns:
    df['currency'] = df['currency'].fillna("INR")

if 'Amount' in df.columns:
    df['Amount'] = df['Amount'].fillna(0)

# ðŸ”¹ 5. Drop duplicates
df = df.drop_duplicates()

# ðŸ”¹ 6. Print summary
print("âœ… Dataset cleaned successfully!\n")
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
```
